## Send and retrieve message with history support

```python
def send_message(message=""):

  specific_message = {"role": "user", "content": message}

  messages.append(specific_message)

  print(messages)

  message = anthropic.messages.create(
    model="claude-3-5-sonnet-20241022",
    max_tokens=1024,
    messages=messages,
  )


  claude_response = {"role": "assistant", "content": message.content[0].text}
  messages.append(claude_response)

  return message.content[0].text
```

```python
print(send_message("What was my last message to you?"))
```
